```python
def operate_a():
        
    def length_c():
        e = input("Enter a string: ")
        print(len(e))
    
    def upperr():
        e = input("Enter a string: ")
        print(e.upper())
    
    def capitalizee():
        e = input("Enter a string: ")
        print(e.capitalize())
    
    def lowerr():
        e = input("Enter a string: ")
        print(e.lower())
    
    def titlee():
        e = input("Enter a string: ")
        print(e.title())
    
    def isalphaa():
        e = input("Enter a string: ")
        print(e.isalpha())
    
    def islowerr():
        e = input("Enter a string: ")
        print(e.islower())
    
    def isupperr():
        e = input("Enter a string: ")
        print(e.isupper())

    def isnumericc():
        e = input("Enter a string: ")
        print(e.isnumeric())
    
    def isspacee():
        e = input("Enter a string: ")
        print(e.isspace())
    
    def choice(i):
        switcher={
            1: length_c,
            2: upperr,
            3: capitalizee,
            4: lowerr,
            5: titlee,
            6: isalphaa,
            7: islowerr,
            8: isupperr,
            9: isnumericc,
            10: isspacee          
        }
        func=switcher.get(i,lambda :'Invalid number.')
        return func()
    
    print("What option would you like to perform?\n\nEnter number corresponding to the operation you would like to perform: ")
    print("\n1: To return length of string.")
    print("2: To return all the letters of string in uppercase.")
    print("3: To capitalize only first letter of string.")
    print("4: To return all the letters of string in lowercase.")
    print("5: To return a title cased string.")
    print("6: To check all the characters of string are alphanumeric or not?")
    print("7: To check all the characters of string are in lowercase or not?")
    print("8: To check all the characters of string are in uppercase or not?")
    print("9: To check all the characters of string are numbers or not?")
    print("10: To check if there are only whitespace characters in the string.\n")
    
    a = int(input("Enter your choice: "))
    choice(a)


def choice_continue():
    j=1
    while(j!=0):
        print("Would you like to perform any operation?")
        c = int(input("Enter your choice (1:yes , 0: no): "))
        if(c==1):
            operate_a()
            j=1
        else:
            j=0
            print("End")
        
choice_continue()
```
